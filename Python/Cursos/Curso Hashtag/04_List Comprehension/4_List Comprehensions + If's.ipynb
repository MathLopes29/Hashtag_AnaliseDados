# List Comprehensions com if para "filtrar" itens

### Estrutura:

- Digamos que eu queira criar uma lista de produtos que bateram a meta.

```python
meta = 1000
vendas_produtos = [1500, 150, 2100, 1950]
produtos = ['vinho', 'cafeiteira', 'microondas', 'iphone']
```

Fazendo por For tradicional

```python
lista_produtos = zip(produtos, vendas_produtos)
resultado = []
for produto, venda in lista_produtos:
    if venda >= meta:
        #print(f'Produto: {produto} com venda de {venda} atingiu a meta!')
        resultado.append(produto)

print(resultado)
```

Fazendo por list comprehension

```python
lista_produtos_atingiu_meta = [produto for produto, venda in zip(produtos, vendas_produtos) if venda >= meta]
print(lista_produtos_atingiu_meta)
```
