```python
vendas_prod = [500, 1000, 850, 200, 100]
nome_prod = ['prodA', 'prodB', 'prodC', 'prodD', 'prodE']

lista = list(zip(nome_prod, vendas_prod))
print(lista)

lista_organizada = sorted(lista, key=lambda x: x[1], reverse=True)
print(lista_organizada)
```

```python
vendas_prod = [500, 1000, 850, 200, 100]
nome_prod = ['prodA', 'prodB', 'prodC', 'prodD', 'prodE']

lista_aux = list(zip(vendas_prod, nome_prod))
lista_aux.sort(reverse=True)
produtos = [produto for venda, produto in lista_aux]
print(produtos)
```
