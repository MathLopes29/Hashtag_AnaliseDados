```python
precos_produtos = [100, 200, 300, 400, 500]
produtos = ['prod1', 'prod2', 'prod3', 'prod4', 'prod5']
impostos = []

for item in precos_produtos:
    impostos.append(item * 0.3)
print(impostos)
```

USANDO LIST COMPREHENSION

```python
impostos = [preco * 0.3 for preco in precos_produtos]
print(impostos)
```

```python
def calcular_impostos(preco, impostos):
    return preco * impostos

impostos = [calcular_impostos(preco, 0.3) for preco in precos_produtos]
print(impostos)
```
