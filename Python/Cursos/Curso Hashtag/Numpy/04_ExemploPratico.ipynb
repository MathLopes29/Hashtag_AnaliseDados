```python
import numpy as np

precos = np.array([100, 250, 40, 20, 35, 99])

novos_precos = precos *1.1
print(novos_precos)
```

```python
soma = np.sum(novos_precos)
print(soma)
```

```python
max_valores = np.max(novos_precos)
print(max_valores)

min_valores = np.min(novos_precos)
print(min_valores)

med_valores = np.mean(novos_precos)
print(med_valores)

desv_padrao = np.std(novos_precos)
print(desv_padrao)

prod_escalar = np.dot(novos_precos, [150, 25, 20, 9, 40, 80])
print(prod_escalar)
```
