#### 23. Faça um Programa para uma loja de tintas. O programa deverá pedir o tamanho em metros quadrados da área a ser pintada. Considere que a cobertura da tinta é de 1 litro para cada 6 metros quadrados e que a tinta é vendida em latas de 18 litros, que custam R\\$ 80,00 ou em galões de 3,6 litros, que custam R\\$ 25,00.

```python
from math import pi


def pintura ():
    cobertura = input("Digite o tamanho da cobertura a ser pintada (em m²): ")
    cobertura = float(cobertura)

    if cobertura <= 0:
        print("Valor inválido. A cobertura deve ser maior que 0 m².")
        return
    
    print(f"Você informou que a cobertura tem {cobertura} m².")
    
    conversao = cobertura/6
    # print(f"Você precisará de {conversao:.2f} L de tinta para pintar {cobertura} m².")

    latas18 = cobertura/18
    # print(f"Você precisará de {latas18:.2f} latas de tinta de 18 L para pintar {cobertura} m².")

    latas36 = cobertura/3.6
    # print(f"Você precisará de {latas36:.2f} latas de tinta de 3,6 L para pintar {cobertura} m².")

    if latas18 > int(latas18):
        latas18 = int(latas18) + 1
    else:
        latas18 = int(latas18)

    if latas36 > int(latas36):
        latas36 = int(latas36) + 1
    else:
        latas36 = int(latas36)

    if latas18 < latas36:
        print("É mais econômico comprar latas de 18 L.")
        custo = float(80/18) * conversao
        print(f"O Custo será de R${custo:.2f}")
    else:   
        print("É mais econômico comprar latas de 3,6 L.")
        custo = float(25/3.6) * conversao
        print(f"O Custo será de R${custo:.2f}")

pintura()
```

```python
def validar_pintura():
    cobertura = input("Digite o tamanho da cobertura a ser pintada (em m²): ")
    cobertura = float(cobertura)

    if cobertura <= 0:
        print("Valor inválido. A cobertura deve ser maior que 0 m².")
        return
    
    print(f"Você informou que a cobertura tem {cobertura} m².")
    
    conversao_litros = cobertura/6
    print(f"Você precisará de {conversao_litros:.2f} L de tinta para pintar {cobertura} m².")

    latas = conversao_litros/18

    if latas > int(latas):
        latas = int(latas) + 1

    print(f"Você precisará de {latas:.2f} latas de tinta de 18 L para pintar {cobertura} m².")

    custo = latas * 80
    print(f"O Custo será de R${custo:.2f}")

validar_pintura()
```

```python
def info_area():
    area = float(input("Digite o tamanho da área a ser pintada (em m²): "))

    if area <= 0:
        print("Valor inválido. A área deve ser maior que 0 m².")
        return
    
    print(f"Você informou que a área tem {area} m².")
    return area

def calcular_tinta(area):
    litros_necessarios = area / 6
    print(f"Você precisará de {litros_necessarios:.2f} L de tinta para pintar {area} m².")
    return litros_necessarios

def calcular_latas(litros):
    latas = litros / 18
    if latas > int(latas):
        latas = int(latas) + 1
    else:
        latas = int(latas)
    print(f"Você precisará de {latas} latas de tinta de 18 L.")
    return latas



def calcular_custo(latas):
    custo_latas = latas * 80
    custo_galao = latas * 25
    custo = custo_galao + custo_latas
    print(f"O custo total será de R${custo:.2f}.")
    return custo
```

```python
area = info_area()
litros = calcular_tinta(area)

def calcular_qtd_latas_galoes(litros):
    latas = 0
    galoes = 0

    
    latas_int = int(litros / 18)
    litros_restantes = litros % 18
    custo_extra_lata = 80

    galoes_int = int(litros_restantes / 3.6)
    if galoes > int(galoes):
        galoes = int(galoes) + 1
    custo_extra_galao = 25 * galoes

    if custo_extra_lata < custo_extra_galao:
        latas =  latas_int + 1
        galoes = 0
    else:
        latas = latas_int
        galoes = galoes_int
        if litros_restantes % 3.6 != 0:
            galoes += 1

    return latas, galoes

latas, galoes = calcular_qtd_latas_galoes(litros)
```
