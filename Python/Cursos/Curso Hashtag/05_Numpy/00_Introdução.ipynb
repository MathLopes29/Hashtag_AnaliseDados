```python
# Lista é uma estrutura de dados que armazena múltiplos valores em uma única variável, porém em análise de dados usamos muito mais arrays pois eles são mais performáticos e otimizadores para operações matemáticas e científicas.

import numpy as np

variavel = np.array([1,2,3,4,5])
print(variavel)

# A diferencial do array para a lista é que o array trabalha com dados homogênios, ou seja, do mesmo tipo!

var_mista = np.array([1, 'abc', 3, 4, 5])
print(type(var_mista))

for elemento in var_mista:
    print(type(elemento))
```
