## Calculando a idade

Um usuário fornece sua data de nascimento no formato "dd/mm/aaaa". Crie um script Python que calcula a idade do usuário.


```python
import datetime

data_atual = datetime.datetime.now(datetime.timezone.utc)
print(f"Data Atual: {data_atual:%d/%m/%Y}")

data_nasc = input("Insira sua data de nascimento (DD/MM/AAAA): ")

data_nasc = datetime.datetime.strptime(data_nasc, "%d/%m/%Y").date()
print(f"Sua data de nascimento é: {data_nasc:%d/%m/%Y}")

idade = data_atual.year - data_nasc.year
if (data_atual.month, data_atual.day)<(data_nasc.month, data_nasc.day):
    idade -= 1

print(f"Sua idade é {idade}")
```
