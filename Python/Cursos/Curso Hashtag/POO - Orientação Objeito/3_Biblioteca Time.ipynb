```python
import time
'''
def wait(seconds):
    """Pause the program for a specified number of seconds."""
    time.sleep(seconds)
    print(f"Waited for {seconds} seconds.")
'''
def hoje():
    """Return the current date in DD/MM/YYYY format."""
    return time.strftime("%d/%m/%Y")

def defina_tempo():
    for i in range(5):
        print(f"Current time: {time.ctime()}")
        time.sleep(1)

def tempo_decorrido():
    for i in range(5):
        tempo_inicial = time.time()
        time.sleep(1)
        tempo_final = time.time()
        print(f"Tempo decorrido: {tempo_final - tempo_inicial} segundos")

if __name__ == "__main__":
    # wait(5)  # Example usage: wait for 5 seconds
    # print(hoje())  # Example usage: print today's date
    # defina_tempo()  # Example usage: print current time every second
    tempo_decorrido()  # Example usage: measure elapsed time over 5 seconds
```

```python
hora_geral = time.gmtime()
# time.localtime()
print(f"Hora geral (UTC): {hora_geral}")
```

```python
dia = hora_geral.tm_mday
mes = hora_geral.tm_mon
ano = hora_geral.tm_year

print(f"Data de Hoje é {dia}/{mes}/{ano}")
```
