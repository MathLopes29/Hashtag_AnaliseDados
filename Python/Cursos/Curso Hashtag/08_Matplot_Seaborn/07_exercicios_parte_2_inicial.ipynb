## Exercício para os três tipos básicos de gráficos

Você recebeu, na forma de listas, os dados de receita anual de uma empresa de tecnologia, em bilhões de dólares, nos últimos anos. Com esses dados, você deve criar três gráficos, um de cada tipo abaixo listado, para visualizar esses dados. Os gráficos devem conter título e rótulos dos eixos. O nome da empresa é "Apple Inc.".

Tipos de gráficos:

- Gráfico de linhas
  - além da linha, mostre os pontos de cada valor
- Gráfico de barras
  - adicione o valor de cada barra no gráfico
- Gráfico de dispersão
  - ligue a grade do gráfico

Cada gráfico deve usar uma paleta de cores diferente.

Os dados são:

Ano | Receita (em bilhões de dólares)
--- | ---
| 2009 | 42,9
| 2010 | 65,2
| 2011 | 108,2
| 2012 | 156,5
| 2013 | 170,9
| 2014 | 182,8
| 2015 | 233,7
| 2016 | 215,6
| 2017 | 229,2
| 2018 | 265,6
| 2019 | 260,2
| 2020 | 274,5
| 2021 | 365,8
| 2022 | 394,3

Qual o tipo de gráfico mais adequado para visualizar esses dados?

```python
# convertendo os dados para listas
import matplotlib.pyplot as plt
# import pandas as pd

ano = [2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
receita = [42.9, 65.2, 108.2, 156.5, 170.9, 182.8, 233.7, 215.6, 229.2, 265.6, 260.2, 274.5, 365.8, 394.3]
```

```python
# gráfico de barras
from matplotlib.lines import lineStyles

cores = plt.get_cmap('Set3').colors


plt.title("Receita Anual da Empresa 2009-2022")
plt.bar(ano, receita, color=cores[6], width=0.5, align='center', edgecolor='black', linewidth=1, alpha=0.7)
plt.xlabel("Ano")
plt.ylabel("Receita em Milhões (R$)")
plt.xticks(ano, rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
from tkinter import font
from cycler import cycler

cores = plt.get_cmap('Set3').colors
ciclos_cores = cycler(color=cores)
plt.rc('axes', prop_cycle = ciclos_cores)

plt.title("Receita Anual da Empresa 2009-2022")
barras = plt.bar(ano, receita, color=cores, width=0.5, align='center', edgecolor='black', linewidth=1, alpha=0.7)
plt.bar_label(barras, labels=[f'R${r}M' for r in receita], padding=3, fontsize=5)
plt.xlabel("Ano", fontsize=8)
plt.ylabel("Receita em Milhões (R$)", fontsize=8)
plt.xticks(ano, rotation=45, fontsize=7)
plt.yticks(fontsize=7)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10,6))

barras = plt.bar(ano, receita, align='center', color='skyblue', edgecolor='black')
plt.bar_label(barras, labels=[f'R${r}M' for r in receita], padding=3, fontsize=8)
plt.xlabel("Ano", fontsize=10)
plt.ylabel("Receita em Milhões (R$)", fontsize=10)
plt.xticks(ano, rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
