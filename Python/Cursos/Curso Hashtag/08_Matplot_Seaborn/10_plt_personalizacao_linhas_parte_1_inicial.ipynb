## Gráfico de linha para apresentação


```python
import matplotlib.pyplot as plt
import numpy as np
from cycler import cycler

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

plt.plot(meses, vendas, marker='o', linestyle='--', linewidth=2)
plt.title('Crescimento das Vendas ao Longo do Ano', fontsize=12, fontweight='bold')

plt.annotate('Menor Valor', color='red', xy=(2, 78), xytext=(3, 78), arrowprops=dict(facecolor='red', arrowstyle='->'), fontsize=8)
plt.annotate('Maior Valor', color='green', xy=(11, 108), xytext=(10, 100), arrowprops=dict(facecolor='green', arrowstyle='->'), fontsize=8)

plt.xlabel('Mês', fontsize=8)
plt.xticks(rotation=45, fontsize=7)
plt.ylabel('Vendas (em milhares)', fontsize=8)
plt.yticks(fontsize=7)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]
marcadores = np.where(vendas == np.min(vendas), 'X', '')

cores = plt.get_cmap('Pastel2').colors
ciclos_cores = cycler('color', cores)
plt.rc('axes', prop_cycle = ciclos_cores)

plt.plot(meses, vendas, linestyle='--', linewidth=2, color='gray')

for posicao, marcador in enumerate(marcadores):
    if marcador == 'X':
        cor='red'
        tamanho = 80
    else:
        cor='green'
        tamanho = 0
    plt.scatter(meses[posicao], vendas[posicao], marker = marcador, color = cor)
    plt.annotate(vendas[posicao], color=cor, xy=(meses[posicao], vendas[posicao]), xytext=(-5, 5), fontsize=8, textcoords='offset points')


plt.axvspan(6, 11, alpha=0.2, color=cores[0])
plt.text(6, 80, "Recuperação de Crescimento", fontsize=8, fontweight='bold', color='green')
plt.box(False)
plt.tick_params(axis='x', length=0)

#plt.xlabel('Mês', fontsize=8)
plt.xticks(rotation=45, fontsize=7)
#plt.ylabel('Vendas (em milhares)', fontsize=8)
plt.yticks(fontsize=7)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
