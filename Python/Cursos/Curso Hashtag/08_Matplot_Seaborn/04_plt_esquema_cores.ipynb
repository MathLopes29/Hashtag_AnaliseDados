## Gráfico de pizza para mostrar a distribuição de categorias de produtos em uma loja

Suponha que você tenha uma loja e queira mostrar a distribuição de categorias de produtos em sua loja. Você pode usar um gráfico de pizza para mostrar a distribuição de categorias de produtos em sua loja:


```python
import matplotlib.pyplot as plt

categorias = ['Eletrônicos', 'Roupas', 'Alimentos', 'Livros']
quantidades = [45, 30, 15, 10]  # valores em porcentagem
destaque = [0.1, 0, 0, 0] # destaque o maior valor
```

```python
plt.pie(quantidades, labels=categorias, startangle=140, explode=destaque, shadow=True, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title('Distribuição de vendas por categoria')
#plt.legend(categorias, title="Categorias de Produtos", loc='upper center')
plt.figure(figsize=(8, 6))
plt.show()
```

O `autopct` é usado para mostrar os valores percentuais. O `%1.1f%%` é usado para mostrar um dígito após o ponto decimal.

```python
plt.pie(quantidades, labels=categorias, autopct='%1.1f%%', startangle=140, explode=destaque, shadow=True, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title('Distribuição de vendas por categoria')
#plt.legend(categorias, title="Categorias de Produtos", loc='upper center')
plt.figure(figsize=(8, 6))
plt.show()
```

Mudando o esquema de cores: https://matplotlib.org/stable/tutorials/colors/colormaps.html

```python
cores = plt.get_cmap('Pastel2').colors

plt.pie(quantidades, labels=categorias, autopct='%1.1f%%', startangle=140, explode=destaque, shadow=True, colors=cores)
plt.title('Distribuição de vendas por categoria')
#plt.legend(categorias, title="Categorias de Produtos", loc='upper center')
plt.figure(figsize=(8, 6))
plt.show()
```
