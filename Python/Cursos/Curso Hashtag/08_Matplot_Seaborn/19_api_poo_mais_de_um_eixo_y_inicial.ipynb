# Trabalhando com eixos



## Mais de um eixo y em uma figura

Crie uma figura de um gráfico de precipitação com um eixo secundário para a temperatura, para mostrar a relação entre a temperatura e a precipitação em uma cidade.



```python
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
precipitacao_media = [200, 180, 160, 100, 60, 30, 30, 35, 60, 100, 160, 200]
temperatura_media = [30, 29, 27, 25, 23, 22, 22, 23, 24, 25, 26, 28]
```

```python
from cProfile import label
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots()
cor_temp = 'tab:red'
cor_prep = 'tab:blue'


ax1.bar(meses, precipitacao_media, color=cor_prep)
ax1.set_ylabel("Precipitação Média (mm)", color=cor_prep)
ax1.tick_params(axis='y', labelcolor=cor_prep)
#ax1.set_ylim(10,200)

ax2 = ax1.twinx()
ax2.plot(meses, temperatura_media, color=cor_temp)
ax2.set_ylabel("Temperatura Média (Cº)", color=cor_temp)
ax2.tick_params(axis='y', labelcolor=cor_temp)
ax2.set_ylim(20,30)

ax1.set_title("Temperatura média e precipitação ao longo do ano")
fig.legend(bbox_to_anchor=(1,1), bbox_transform=ax1.transAxes)

plt.show()
```
