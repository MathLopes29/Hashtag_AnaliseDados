# Trabalhando com eixos



## Gráfico de linhas

```python
import matplotlib.pyplot as plt

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]
```

```python
fig, ax = plt.subplots()

ax.plot(meses, vendas, linestyle='--', marker='o', color='green', linewidth=2)
plt.box(False)
plt.xticks(fontsize=8)
plt.yticks([])
ax.tick_params('x', length=0)
plt.show()
```

## Gráfico de barras

```python
import matplotlib.pyplot as plt

produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [23, 45, 17, 30, 37]
```

```python
fig, ax = plt.subplots()

ax.bar(produtos, quantidades)
plt.show()
```

## Gráfico de pizza

```python
categorias = ['Eletrônicos', 'Roupas', 'Alimentos', 'Livros']
quantidades = [45, 30, 15, 10]
destaque = [0.1, 0, 0, 0]  # destacar a primeira fatia
```

```python
fig, ax = plt.subplots()

ax.pie(quantidades, labels=categorias, autopct='%1.1f%%', explode=destaque, shadow=True, startangle=45)
ax.axis('equal')
plt.show()
```

## Gráfico de barras empilhadas

```python
anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20]
vendas_loja = [10, 8, 6, 4]
```

```python
barra_online = plt.bar(anos, vendas_online, label='Online')
barra_loja = plt.bar(anos, vendas_loja, bottom=vendas_online, label='Loja')

plt.bar_label(barra_online, labels=vendas_online, padding=3, fontsize=8, fontweight='bold', label_type='center')
plt.bar_label(barra_loja, labels=vendas_loja, padding=3, fontsize=8, fontweight='bold', label_type='center')

plt.xticks(ticks=anos)
plt.yticks([])
plt.legend()
plt.box(False)
#plt.grid(axis='y', linestyle='--', alpha=0.8)
plt.tick_params(axis='x', length=0)
plt.tick_params(axis='y', length=0)
```
