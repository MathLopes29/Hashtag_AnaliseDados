Construa um gráfico de linhas com as seguintes características:

- Tamanho 14x6
- Linha com espessura 3
- Palheta de cores Set3
- Destaque o impacto da pandemia (2020 e 2021) com um vspan
- Destaque o impacto da pandemia com um texto
- Destaque os pontos de início e fim da série com um scatter plot com as quantidades de refeições servidas

- Adicione um título

```python
# listas de anos e refeições
anos = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
refeicoes = [40139, 127020, 168193, 153115, 202102, 232897, 277912, 205350, 233389, 232797, 110023, 160350, 222789]
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from cycler import cycler

cores = plt.get_cmap('Set3').colors
ciclos_cores = cycler('color', cores)
plt.rc('axes', prop_cycle=ciclos_cores)


plt.plot(anos, refeicoes, marker='o', linestyle='--', linewidth=2)
plt.box(False)
plt.title("Evolução Anual de Refeições", fontweight='bold', fontsize=14, loc='center', pad='20')

for p in (-1,0):
    tamanho = 60
    plt.scatter(anos[p], refeicoes[p], marker='o', s=tamanho, color=cores[4], zorder=2)
    plt.annotate(refeicoes[p], color=cores[4], xy=(anos[p], refeicoes[p]), xytext=(5,0), textcoords='offset points', size=9, weight='bold')

plt.yticks([])
plt.axvspan(anos[anos.index(2020)], anos[anos.index(2021)], alpha=0.2, color=cores[1])
plt.text(2020, 40000, 'Pandemia', fontsize=11, color=cores[4], weight='bold')
plt.xticks(anos, size= 11, fontsize=8, rotation=45)
plt.tick_params(axis='x', length=0)
plt.savefig('grafico_linhas.png', bbox_inches='tight', dpi=300)
plt.show()
```
