## Gráfico de barras para apresentação

Considere que você tem as quantidades de produtos vendidos em uma loja. Faça um gráfico de barras horizontais.

```python
import matplotlib.pyplot as plt
from cycler import cycler

cores = plt.get_cmap('Pastel1').colors
ciclos_cores = cycler('color', cores)
plt.rc('axes', prop_cycle=ciclos_cores)

produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [23, 45, 17, 30, 37]
rotulos = [valor if valor == max(quantidades) else '' for valor in quantidades]

barras = plt.barh(produtos, quantidades)
#plt.bar_label(barras, labels=rotulos, padding=3, fontsize=8, fontweight='bold')
plt.bar_label(barras, labels=quantidades, padding=3, fontsize=8, fontweight='bold')
plt.box(False)
plt.xticks([])
plt.tick_params(axis='y', length=0)

for i, barra in enumerate (barras):
    if i != quantidades.index(max(quantidades)):
        barras[i].set_color(cores[6])
        barras[i].set_alpha(0.3)
    else:
        barras[i].set_color(cores[2])
        barras[i].set_alpha(0.3)

plt.show()
```

```python
dados_ordenados = sorted(zip(quantidades, produtos))

quantidades_ordenadas = [dado[0] for dado in dados_ordenados]
print(quantidades_ordenadas)
```

```python
produtos_ordenadas = [dado[1] for dado in dados_ordenados]
print(produtos_ordenadas)
```
