```python
df_franca.set_index('Year', inplace=True)
df_franca.sort_index(inplace=True)
display(df_franca)
```

```python
receita = df_franca.groupby('Year').sum(numeric_only=True)
receita = receita[['US$ FOB']]

def format_valor(x):
    valor = f"{x:,.0f}"
    return valor

receita['US$ FOB'] = receita['US$ FOB'].apply(format_valor)

print(receita)
```

```python
produtos = df_franca.groupby('SH2 Description').sum(numeric_only=True)
produtos = produtos[['US$ FOB']]
display(produtos)
```

```python
produtos = produtos.sort_values('US$ FOB', ascending=False)
produtos_top10 = produtos.head(10)
display(produtos_top10)

produtos_top10.plot.bar()
```

```python
df_franca.reset_index(inplace=True)

ano2020 = df_franca.loc[df_franca['Year'] == 2020]
cidades2020 = ano2020[['City', 'US$ FOB']].groupby('City').sum()
table = cidades2020.sort_values('US$ FOB', ascending=False).head(10)
display(table)
```

```python
produtosmaiores = df_franca[['SH2 Description', 'US$ FOB']].groupby('SH2 Description').sum()
produtosmaiores = produtosmaiores.sort_values(by='US$ FOB', ascending=False).head(10)
produtosmaiores['US$ FOB'] = produtosmaiores['US$ FOB'].apply(format_valor)
display(produtosmaiores)
```
