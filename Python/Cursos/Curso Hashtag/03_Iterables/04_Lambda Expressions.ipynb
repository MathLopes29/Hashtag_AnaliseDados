# Principal Aplicação de Lambda Expressions

### Usar lambda como argumento de alguma outra função, como map e filter

```python
preco_tecnologia = {'notebook asus': 2450, 'iphone': 4500, 'samsung galaxy': 3000, 'tv samsung': 1000, 'ps5': 3000, 'tablet': 1000, 'notebook dell': 3000, 'ipad': 3000, 'tv philco': 800, 'notebook hp': 1700}
```

### map()
- Queremos saber o preço de cada produto adicionando o valor do imposto de 30% sobre o valor do produto

```python
#Fazendo por function

def item(x):
    return x[1] * 0.3 + x[1]

lista = list(map(item, preco_tecnologia.items()))
print(lista)
```

```python
def a(x):
    return x[1] * 0.5 + x[1]

lista = list(map(a, preco_tecnologia.items()))
print(lista)
```

```python
#fazendo por lambda
lista = list(map(lambda x: x[1] * 0.3 + x[1], preco_tecnologia.items()))
print(lista)

lista_final = list(map(lambda x: x[1] * 0.5 + x[1], preco_tecnologia.items()))
lista_final = sorted(lista_final)
print(lista_final)
```

### filter()
- Queremos apenas os produtos que custam acima de 2000

```python
#fazendo por function

#lista_produto = dict(preco_tecnologia.items())
#print(lista_produto)

def condicao(x):
    return x[1] >= 2000

lista_filtrada = list(filter(condicao, preco_tecnologia.items()))
print(lista_filtrada)
```

```python
#fazendo por lambda
lista_filtrada = dict(list(filter(lambda x: x[1]>=2000, preco_tecnologia.items())))
print(lista_filtrada)
```
