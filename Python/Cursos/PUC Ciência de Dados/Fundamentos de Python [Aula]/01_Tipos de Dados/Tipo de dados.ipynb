<font color="#0D9DF0"><h3><B>O QUE SÃO TIPOS DE DADOS EM PYTHON?</b></h3></font>
***


Todo valor em Python possui um tipo. Esses tipos determinam o que você pode fazer com esses valores, como somar, fatiar, iterar, entre outras operações.

Python é uma linguagem de tipagem dinâmica, ou seja, você não precisa declarar o tipo da variável — o próprio Python reconhece automaticamente.


##### Por que é importante saber os tipos? 
Saber os tipos de dados ajuda a:

- Usar corretamente as variáveis

- Fazer operações válidas (ex: não dá pra somar texto com número direto)
 
- Controlar melhor a entrada/saída de dados
 
- Trabalhar com listas, dicionários, e outras estruturas com segurança
 
- Fazer cast/conversões de tipos (int(), str(), etc.)


##### Quais são os tipos mais usados?

<center>

| Tipo de Dado   | Exemplo              | Para que serve?                                      |
|----------------|----------------------|------------------------------------------------------|
| `int`          | `10`, `-3`, `0`      | Números inteiros, usados em contagem, operações     |
| `float`        | `3.14`, `-2.5`, `0.0`| Números decimais, usados para medidas, porcentagens |
| `str`          | `"Olá"` ou `'Python'`| Texto, nomes, mensagens, etc                         |
| `bool`         | `True`, `False`      | Lógica condicional, verificações                    |


</center>


<font color="##188038"><h4><B>  1. Tipo str – Strings (Texto)</b></h4></font>
***
 O que é?
É uma sequência de caracteres (letras, números ou símbolos). Usado para nomes, mensagens, textos, etc.

Como criar:

```python
texto = str("Olá, mundo!") 
print(texto)
print(texto[0])
print(len(texto))
```

##### Funções/métodos úteis:
<center>

| **Função ou método**         | **O que faz**                                   |
|-----------------------------|--------------------------------------------------|
| `len(texto)`                | Retorna o número de caracteres                   |
| `texto.upper()`             | Converte para MAIÚSCULAS                         |
| `texto.lower()`             | Converte para minúsculas                         |
| `texto.strip()`             | Remove espaços extras das bordas                 |
| `texto.replace("a", "b")`   | Substitui `"a"` por `"b"`                        |
| `texto.split()`             | Divide o texto em uma lista de palavras          |

</center>

##### Exemplo prático:

```python
print(texto.lower())
print(texto.split())
```

##### F-strings (formatar string com variáveis):

```python
texto = str("Olá, mundo!")
print(texto + " Bem-vindo ao curso de Python!")
```


<font color="##188038"><h4><B> 2. Tipo int – Números Inteiros</b></h4></font>
***
O que é?
Números sem parte decimal, usados para contagens, idades, etc.

Como criar:

```python
x = int(15)
print(x)
```

```python
y = float(input("Digite um número: "))
print(y*x)
```

```python
idade = int(input("Digite sua idade: "))
if idade < 100 and idade > 0:
    if idade >= 18:
        print("Maior de idade")
    else:
        print("Menor de idade")
else:
    print("Idade inválida")
```

##### Operações básicas:
<center>

| **Operação**         | **Exemplo**      | **Resultado** |
|:--------------------:|:----------------:|:-------------:|
| Soma                 | `5 + 3`          | `8`           |
| Subtração            | `10 - 4`         | `6`           |
| Multiplicação        | `2 * 3`          | `6`           |
| Divisão inteira      | `10 // 3`        | `3`           |
| Resto da divisão     | `10 % 3`         | `1`           |
| Potência             | `2 ** 3`         | `8`           |


##### Exemplo prático:

```python
x = 10
y = 5

print("Adição: ", x+y)
print("Subtração: ", x-y)
print("Multiplicação: ", x*y)
print("Divisão: ", x/y)
print("Divisão inteira: ", x//y)
print("Módulo: ", x%y)
print("Potência: ", x**y)
```

<font color="##188038"><h4><B> 4. Tipo bool – Booleanos (Lógicos)</b></h4></font>
***
O que é?
Representa apenas dois valores:

- `True` (verdadeiro)

 - `False` (falso)

Usado para condições e comparações.

Como criar:

```python
logica = bool(input("Digite um valor booleano: "))
print(logica) # True será qualquer valor diferente de vazio, False será apenas vazio
```

#### Criado geralmente por comparações:

```python
x = str(input("Digite um valor para comparar: "))
valor = "P"
if x == valor:
    print("Valores Iguais!")
else:
    print("Diferentes!")
```
