```python
# Função que categoriza os países das Américas.
def categorizar_america(pais):
    match pais:
        # América do Sul
        case 'Argentina' | 'Bolivia' | 'Brazil' | 'Chile' | 'Colombia' | 'Ecuador' | 'Paraguay' | 'Peru' | 'Uruguay' | 'Venezuela':
            return 'América do Sul'

        # América do Norte
        case 'Canada' | 'United States' | 'Mexico':
            return 'América do Norte'

        # América Central / Caribe
        case 'Costa Rica' | 'Cuba' | 'Dominican Republic' | 'El Salvador' | 'Guatemala' | 'Haiti' | 'Honduras' | 'Jamaica' | 'Nicaragua' | 'Panama' | 'Puerto Rico' | 'Trinidad and Tobago':
            return 'América Central/Caribe'

        case _:
            return 'Outro continente'
        
df_tratado['Região América'] = df_tratado['País'].apply(categorizar_america)
```

```python
# 4ª - Comparação do PIB per capita médio entre as regiões das Américas para um ano específico escolhido pelo usuário.
ano_escolhido = int(input("Digite o ano que deseja visualizar, entre 1952 e 2007, em intervalos de 5 anos: "))
if ano_escolhido in [1952, 1957, 1962, 1967, 1972, 1977, 1982, 1987, 1992, 1997, 2002, 2007]:
    print("Ano válido")
else:
    print("Ano inválido. Por favor, escolha um ano entre 1952 e 2007, em intervalos de 5 anos.")

# Filtrar ano e remover outros continentes
filtro = (
    (df_tratado['Ano'] == ano_escolhido) &
    (df_tratado['Região América'] != 'Outro continente')
)

# Agrupar por região
pib_regiao = (
    df_tratado[filtro]
    .groupby('Região América')['Produto Interno Bruto per capita']
    .mean()
)

plt.figure(figsize=(12, 6))

barras = plt.bar(
    pib_regiao.index,
    pib_regiao.values,
    color='skyblue',
    edgecolor='black',
    linewidth=1,
    alpha=0.7
)

plt.bar_label(
    barras,
    labels=[f'$ {valor:.2f}' for valor in pib_regiao.values],
    padding=3,
    fontsize=8
)

plt.title(f' Média do PIB per capita das Américas - {ano_escolhido}', fontsize=12, fontweight='bold')
plt.xlabel('Região', fontsize=10, fontweight='bold')
plt.ylabel('PIB per capita', fontsize=10, fontweight='bold')
plt.xticks(rotation=20)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# 5ª - Comparação da evolução do PIB per capita ao longo do tempo para um conjunto de países escolhidos pelo usuário.
paises = input("Digite os países que deseja comparar, separados por virgula e o nome do país em inglês: ").split(',')
for pais in paises:
    paises = [pais.strip() for pais in paises]
else:    
    print("Países válidos")

df_filtrado = df_tratado[df_tratado['País'].isin(paises)]

fig, ax = plt.subplots(figsize=(12, 8))

for pais in paises:
    dados_pais = df_filtrado[df_filtrado['País'] == pais]
    ax.plot(
        dados_pais['Ano'],
        dados_pais['Produto Interno Bruto per capita'],
        marker='o',
        label=pais
    )

ax.set_title('PIB per capita ao longo do tempo', fontsize=12, fontweight='bold')
ax.set_xlabel('Ano', fontsize=10, fontweight='bold', labelpad=10)
ax.set_ylabel('PIB per capita', fontsize=10, fontweight='bold', labelpad=10)
ax.legend()
ax.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
# 6ª - Gráfico de pizza mostrando a distribuição percentual da população por continente para um ano específico escolhido pelo usuário.
from cycler import cycler

#cores = plt.get_cmap('Set3').colors
#ciclo_cores = cycler(color=cores)
#plt.rc('axes', prop_cycle=ciclo_cores)

ano_escolhido = int(input("Digite o ano que deseja visualizar, entre 1952 e 2007, em intervalos de 5 anos: "))
dados = df_tratado[df_tratado['Ano'] == ano_escolhido].groupby('Continente')['População total'].sum()
maior_continente = dados.idxmax()
destaque = [0.1 if continente == maior_continente else 0 for continente in dados.index]

plt.pie(
    dados,
    labels=dados.index,
    autopct='%1.1f%%',
    startangle=140,
    shadow=True, 
    explode=destaque
)

plt.title(f'Distribuição da População por Continente - {ano_escolhido}', fontsize=10, fontweight='bold', pad=20)
plt.axis('equal')
plt.show()
```

```python
# 7ª - Gráfico de Pizza mostrando a distribuição de Países por Condições de Expectativa de Vida (Alta, Média, Baixa) para um ano específico escolhido pelo usuário.
from cycler import cycler

#cores = plt.get_cmap('Set3').colors
#ciclo_cores = cycler(color=cores)
#plt.rc('axes', prop_cycle=ciclo_cores)

ano_escolhido = int(input("Digite o ano que deseja visualizar, entre 1952 e 2007, em intervalos de 5 anos: "))
dados = df_tratado[df_tratado['Ano'] == ano_escolhido].groupby('Condição')['Condição'].count()
maior_continente = dados.idxmax()
destaque = [0.1 if continente == maior_continente else 0 for continente in dados.index]

plt.pie(
    dados,
    labels=dados.index,
    autopct='%1.1f%%',
    startangle=140,
    shadow=True, 
    explode=destaque
)

plt.title(f'Distribuição por Condições de Expectativa de Vida - {ano_escolhido}', fontsize=10, fontweight='bold', pad=20)
plt.axis('equal')
plt.show()
```
