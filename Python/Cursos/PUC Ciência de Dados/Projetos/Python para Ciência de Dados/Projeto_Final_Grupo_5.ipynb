```python
from tkinter import font
from cycler import cycler

continente = df_tratado['Continente'].unique()
expectativa_media = df_tratado['Expectativa de vida'].groupby(df_tratado['Continente']).mean()


#cores = plt.get_cmap('Set3').colors
#ciclos_cores = cycler(color=cores)
#plt.rc('axes', prop_cycle = ciclos_cores)

plt.figure(figsize=(12, 6))
plt.bar(continente,expectativa_media)
plt.title('Expectativa de Vida Média por Continente', fontsize=10, fontweight='bold')
barras = plt.bar(continente, expectativa_media, color='SkyBlue', align='center', edgecolor='black', linewidth=1, alpha=0.5)
plt.bar_label(barras, labels=[f'{r:.2f} Anos' for r in expectativa_media], padding=3, fontsize=7, color='black', fontweight='bold')
plt.xlabel("Continente", fontsize=8, style='italic')
plt.ylabel("Expectativa de Vida", fontsize=8, style='italic')
plt.xticks(continente, rotation=45, fontsize=7)
plt.yticks(fontsize=7)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
evolucao = df_tratado.groupby(['Ano', 'Continente'])['População total'].sum().unstack()

plt.figure(figsize=(12, 6))

for continente in evolucao.columns:
    plt.plot(evolucao.index, evolucao[continente], marker='o', linestyle='--', label=continente)

plt.title('Evolução da população por continente', fontsize=15, fontweight='bold')
plt.xlabel('Ano', fontsize=9, fontweight='bold')
plt.xticks(evolucao.index, rotation=45, fontsize=7)
plt.ylabel('População total', fontsize=9, fontweight='bold')
plt.yticks(fontsize=7)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Continente', fontsize=7, title_fontsize=8)
plt.tight_layout()
plt.show()
```

```python
def categorizar_america(pais):
    match pais:
        # América do Sul
        case 'Argentina' | 'Bolivia' | 'Brazil' | 'Chile' | 'Colombia' | 'Ecuador' | 'Paraguay' | 'Peru' | 'Uruguay' | 'Venezuela':
            return 'América do Sul'

        # América do Norte
        case 'Canada' | 'United States' | 'Mexico':
            return 'América do Norte'

        # América Central / Caribe
        case 'Costa Rica' | 'Cuba' | 'Dominican Republic' | 'El Salvador' | 'Guatemala' | 'Haiti' | 'Honduras' | 'Jamaica' | 'Nicaragua' | 'Panama' | 'Puerto Rico' | 'Trinidad and Tobago':
            return 'América Central/Caribe'

        case _:
            return 'Outro continente'
        
df_tratado['Região América'] = df_tratado['País'].apply(categorizar_america)
```

```python
df_tratado['Região América'].value_counts()
```

```python
ano_escolhido = int(input("Digite o ano que deseja visualizar: "))
match ano_escolhido:
    case 1952, 1957, 1962, 1967, 1972, 1977, 1982, 1987, 1992, 1997, 2002, 2007:
        print("Ano válido")
    case _:
        print("Ano invalido. Por favor, escolha um ano entre 1952 e 2007, em intervalos de 5 anos.")
    
# filtrar ano e remover outros continentes
filtro = (
    (df_tratado['Ano'] == ano_escolhido) &
    (df_tratado['Região América'] != 'Outro continente')
)

# agrupar por região
pib_regiao = (
    df_tratado[filtro]
    .groupby('Região América')['Produto Interno Bruto per capita']
    .mean()
)

plt.figure(figsize=(12, 6))

barras = plt.bar(
    pib_regiao.index,
    pib_regiao.values,
    color='skyblue',
    edgecolor='black',
    linewidth=1,
    alpha=0.7
)

plt.bar_label(
    barras,
    labels=[f'$ {valor:.2f}' for valor in pib_regiao.values],
    padding=3,
    fontsize=8
)

plt.title(f' Média do PIB per capita das Américas - {ano_escolhido}', fontsize=12, fontweight='bold')
plt.xlabel('Região', fontsize=10, fontweight='bold')
plt.ylabel('PIB per capita', fontsize=10, fontweight='bold')
plt.xticks(rotation=20)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
paises = ['Brazil', 'Italy', 'Japan', 'France', 'United States']
df_filtrado = df_tratado[df_tratado['País'].isin(paises)]

fig, ax = plt.subplots(figsize=(12, 8))

for pais in paises:
    dados_pais = df_filtrado[df_filtrado['País'] == pais]
    ax.plot(
        dados_pais['Ano'],
        dados_pais['Produto Interno Bruto per capita'],
        marker='o',
        label=pais
    )

ax.set_title('PIB per capita ao longo do tempo', fontsize=12, fontweight='bold')
ax.set_xlabel('Ano', fontsize=10, fontweight='bold')
ax.set_ylabel('PIB per capita', fontsize=10, fontweight='bold')
ax.legend()
ax.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
