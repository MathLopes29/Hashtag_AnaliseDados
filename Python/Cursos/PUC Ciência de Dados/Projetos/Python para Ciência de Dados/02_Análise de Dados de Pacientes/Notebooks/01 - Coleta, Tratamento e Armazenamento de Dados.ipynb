#### Corrigindo dados ausentes
1. Média para as variáveis numéricas 
1. Moda para as variáveis categóricas

```python
df_clinicos['peso'].fillna(df_clinicos['peso'].median().round(2), inplace=True) 
#df_clinicos['genero'].fillna(df_clinicos['genero'].mode()[0], inplace=True)
df_clinicos['genero'].fillna("Não Informado", inplace=True)
```

```python
df_clinicos.isnull().sum()
```

```python
df_pacientes['qtde_filhos'].fillna(df_pacientes['qtde_filhos'].median().round(0), inplace=True)
#df_pacientes['classe_trabalho'].fillna(df_pacientes['classe_trabalho'].mode()[0], inplace=True)
df_pacientes['classe_trabalho'].fillna("Não Informado", inplace=True)
```

```python
df_pacientes.isnull().sum()
```

```python
df_pacientes['idade'] = pd.to_numeric(df_pacientes['idade'], errors='coerce')
df_pacientes['qtde_filhos'] = pd.to_numeric(df_pacientes['qtde_filhos'], errors='coerce')
df_pacientes['estado_civil'] = df_pacientes['estado_civil'].astype('string')
df_pacientes['raca'] = df_pacientes['raca'].astype('string')
df_pacientes['classe_trabalho'] = df_pacientes['classe_trabalho'].astype('string')
df_pacientes['escolaridade'] = df_pacientes['escolaridade'].astype('string')
```

#### 5. Integração de Dados
A integração de dados é um processo essencial em projetos de análise, especialmente quando os dados vêm de fontes diferentes. Muitas vezes, informações importantes estão dispersas em múltiplos bancos de dados, planilhas, APIs ou sistemas, e cada fonte pode conter uma parte relevante do contexto necessário para uma análise completa. A integração permite combinar essas fontes em uma única visão consolidada, tornando possível analisar o conjunto de dados de forma mais rica e abrangente.
***

```python
df_pacientes.head(5)
```

```python
df_integrado = df_pacientes.merge(df_clinicos, on='id_cliente', how='inner').merge(df_região, left_on='id_estado', right_on='id_estado', how='inner')
display(df_integrado.head(5))
```

```python
df_integrado_copy = df_integrado[['id_cliente', 'idade', 'genero', 'estado_civil', 'peso', 'colesterol', 'id_estado', 'sigla', 'estado', 'regiao']]
display(df_integrado_copy.head(5))
```
