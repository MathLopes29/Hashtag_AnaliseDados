<font color="#0D9DF0"><h5>Visualizações Gráficas</h5></font>
Análises gráficas são fundamentais no contexto de análise de dados porque permitem uma interpretação rápida e intuitiva de padrões, tendências e relações entre variáveis, algo que muitas vezes não fica claro ao observar apenas tabelas ou números. Gráficos têm a capacidade de transformar dados brutos em histórias visuais, ajudando a identificar insights, como relações de causa e efeito, correlações e outliers, que podem impactar as decisões estratégicas.
***



<font color="0098a6"><h5>Histograma</h5></font>
Mostram a frequência de valores de uma variável, permitindo observar a forma da distribuição, como se é simétrica ou assimétrica.


<font color="0098a6"><h5>Boxplots</h5></font>


<font color="#0D9DF0"><h5>Análise de Correlação</h5></font>
A correlação mede o grau de associação linear entre variáveis numéricas, variando de -1 a +1.
- Valores próximos de +1 indicam uma forte correlação positiva 
    - (quando uma variável aumenta, a outra tende a aumentar) 
- Valores próximos de -1 indicam uma correlação negativa. 
- Valores em torno de 0 indicam pouca ou nenhuma correlação linear.
***


<font color="#0D9DF0"><h5>Heatmap</h5></font>

Para visualizar a correlação entre variáveis, é comum usar um heatmap de correlação, que apresenta as correlações em forma de matriz e facilita a identificação de pares de variáveis altamente correlacionadas, ajudando a orientar decisões como a eliminação de variáveis redundantes.
***


<font color="#0D9DF0"><h5>Dispersão</h5></font>
O gráfico de dispersão, ou scatter plot, é uma ótima ferramenta para visualizar a relação entre duas variáveis numéricas. Cada ponto no gráfico representa uma observação do conjunto de dados, onde a posição horizontal (eixo X) e vertical (eixo Y) mostram os valores de cada variável. Esse tipo de gráfico é útil para identificar tendências, padrões, outliers e correlações entre variáveis.

https://leandrolessa.com.br/tutoriais/grafico-de-dispersao-como-criar-e-analisar-na-pratica/


<font color="#0D9DF0"><h5>Barras</h5></font>

O gráfico de barras é uma ótima escolha para visualizar variáveis categóricas, pois representa a contagem ou a média de valores de cada categoria. Cada barra do gráfico indica a frequência de uma categoria específica, o que facilita a comparação entre diferentes grupos. Esse tipo de gráfico é útil para entender como os dados estão distribuídos em variáveis como estado_civil, classe_trabalho e regiao.

<font color="#0D9DF0"><h5>Agrupamentos</h5></font>
Realizar agrupamentos em conjuntos de dados é uma etapa fundamental na análise de dados, pois permite entender como diferentes subgrupos se comportam dentro de um conjunto maior. Ao agrupar os dados, é possível identificar padrões, tendências e outliers que poderiam passar despercebidos em uma análise geral. Além disso, os agrupamentos ajudam a segmentar os dados para realizar comparações mais precisas entre diferentes categorias, como faixas etárias, classes sociais, regiões geográficas ou qualquer outro critério relevante.
