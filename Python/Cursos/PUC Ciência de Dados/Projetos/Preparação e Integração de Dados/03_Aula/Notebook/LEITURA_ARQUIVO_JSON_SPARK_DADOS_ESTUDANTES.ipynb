# IMPORTAÇÃO DAS BIBLIOTECAS

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import to_timestamp, col
```

# CRIAÇÃO DA APLICAÇÃO SPARK

```python
spark = SparkSession.builder.getOrCreate()
```

# VARIÁVEIS DE APOIO

```python
arquivo = '/content/DADOS_ESTUDANTES.json'

options = {
    "encoding": "utf-8",
    "inferSchema": "true",
    "multiline": "true"
}
```

# LEITURA DO ARQUIVO JSON USANDO SPARK

```python
df = spark.read.options(**options).json(arquivo)
```

# EXIBINDO UMA AMOSTRA DOS DADOS

```python
df.show()
```

# EXIBINDO OS METADADOS (SCHEMA) DO ARQUIVO

```python
df.printSchema()
```

# AJUSTAR O SCHEMA DOS DADOS, SE NECESSÁRIO

```python
df = df.withColumn(
    "Data de Ingresso",
    to_timestamp(col("Data de Ingresso"), "dd/MM/yyyy")
)

df = df.withColumn(
    "Previsão de Formatura",
    to_timestamp(col("Previsão de Formatura"), "dd/MM/yyyy")
)

df = df.withColumn("Média de Notas", col("Média de Notas").cast("double"))
df = df.withColumn("Matrícula", col("Matrícula").cast("int"))
```
