## IMPORTAÇÃO DAS BIBLIOTECAS

```python
from azure.storage.blob import BlobClient
import os
```

## MÉTODO PARA CARREGAR ARQUIVOS NO AZURE STORAGE BLOB

```python
def carrega_arquivo_datalake(diretorio_local, diretorio_datalake, arquivo, container, conexao):
    for arquivo in os.listdir(diretorio_local):
        arquivo_blob = diretorio_datalake + '/' + os.path.splitext(os.path.basename(arquivo))[0] + '/' + arquivo
        blob = BlobClient.from_connection_string(conn_str=conexao, container_name = container, blob_name = arquivo_blob)
        with open(diretorio_local + arquivo, "rb") as data:
            blob.upload_blob(data, overwrite = True)
```
