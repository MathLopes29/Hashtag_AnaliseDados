## IMPORTAÇÃO DAS BIBLIOTECAS

```python
import boto3
import os
```

## MÉTODO PARA CARREGAR ARQUIVOS NO AZURE STORAGE BLOB

```python
def carrega_arquivo_minio(diretorio_local, diretorio_datalake, arquivo, bucket, endpoint, access_key, secret_key):
    s3 = boto3.client(
        's3',
        endpoint_url=endpoint,
        aws_access_key_id=access_key,
        aws_secret_access_key=secret_key
    )
    caminho_arquivo = os.path.join(diretorio_local, arquivo)
    objeto_s3 = f"{diretorio_datalake}/{os.path.splitext(arquivo)[0]}/{arquivo}"
    s3.upload_file(
        caminho_arquivo,
        bucket,
        objeto_s3
    )
    print(f"Arquivo enviado: {objeto_s3}")
```

## PROCEDIMENTO PRINCIPAL

```python
bucket = 'datalake'
endpoint = "http://localhost:9000"
access_key = "minioadmin"
secret_key = "minioadmin"
diretorio_datalake = "bronze"
diretorio_local = "COLE_AQUI_SEU_DIRETORIO"
lista_arquivos = [
    "DADOS_ESTUDANTES.json",
    "DADOS_VOOS.parquet",
    "DADOS_EXAMES.csv",
    "DADOS_ALUNOS.xml",
    "DADOS_BANCARIOS.xml"
]

for arquivo in lista_arquivos:
    carrega_arquivo_minio(
        diretorio_local,
        diretorio_datalake,
        arquivo,
        bucket,
        endpoint,
        access_key,
        secret_key
    )
```
