#### Estruturas Condicionais (DecisĂµes)
***
As estruturas condicionais permitem que o programa tome decisĂµes com base em uma condiĂ§ĂŁo. Usamos:

`if`: se a condiĂ§ĂŁo for verdadeira

`elif`: senĂŁo se...

`else`: senĂŁo (caso contrĂˇrio)

```python
idade = int(input("Digite sua idade:"))
if idade >= 18:
    print("VocĂŞ Ă© maior de idade")
elif idade > 100:
    print("Idade invĂˇlida")
else:
    print("VocĂŞ Ă© menor de idade")
    if idade < 0:
        print("Idade invĂˇlida")
```

#### Estruturas de RepetiĂ§ĂŁo (LaĂ§os de Loop)
***
Permitem executar blocos de cĂłdigo repetidamente. Os dois principais em Python sĂŁo:

`for`: para um nĂşmero definido de repetiĂ§Ăµes

`while`: enquanto uma condiĂ§ĂŁo for verdadeira

##### Exemplo com `for`:

```python
lista = [1,2,3,4,5]
for i in lista:
    print(i)
```

##### Exemplo com `while`:

```python
contador = 0
while True:
    contador += 1
    if contador == 15:
        break
    print(contador)
```

#### Comandos de Controle de Fluxo
***
Controlam o comportamento dentro dos loops.

`break`: encerra o loop imediatamente

`continue`: pula para a prĂłxima iteraĂ§ĂŁo

`pass`: faz nada (placeholder)

Exemplo com `break` e `continue`

```python
lista_senhas = ['a','b','c','d']

for i in lista_senhas:
    if i == 'c':
        print('Aprovado!')
        continue
    else:
        print('Reprovado!')
        
```
