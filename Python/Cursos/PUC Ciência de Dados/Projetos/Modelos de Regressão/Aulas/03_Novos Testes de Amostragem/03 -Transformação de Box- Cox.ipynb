## Transformação Box-Cox

```python
import numpy as np
import pandas as pd
from scipy.stats import boxcox
import matplotlib.pyplot as plt
```

#### 1. Geração de Dados: 
Usamos uma distribuição exponencial para simular dados assimétricos (simulando dados não normalizados)

```python
np.random.seed(42)
data = np.random.exponential(scale=2, size=1000)
df = pd.DataFrame({'dados_originais': data})
```

```python
df
```

#### 2.Realizando a transformação de Box-Cox
A transformação de Box-Cox requer que todos os valores sejam positivos

```python
transformed_data, lambda_value = boxcox(df['dados_originais'])
```

```python
# Adicionando a coluna transformada ao DataFrame
df['dados_transformados'] = transformed_data
```

```python
transformed_data
```

```python
# Exibindo o lambda estimado
print(f"Valor de lambda estimado: {lambda_value:.4f}")
```

```python
# Plotando antes e depois da transformação
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(df['dados_originais'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribuição Original')
plt.xlabel('Valor')
plt.ylabel('Frequência')

plt.subplot(1, 2, 2)
plt.hist(df['dados_transformados'], bins=30, color='lightcoral', edgecolor='black')
plt.title('Distribuição Após Box-Cox')
plt.xlabel('Valor')
plt.ylabel('Frequência')

plt.tight_layout()
plt.show()
```
